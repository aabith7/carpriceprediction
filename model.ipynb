```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import pickle
```

```python
df = pd.read_csv('car_price_data_large.csv')
df.head()
df
```

```python
x=df[['Year','Engine_Size','Mileage']]
```

```python
x
```

```python
y = df['Price']
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled_train = scaler.fit_transform(x_train)
x_scaled_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()

regression.fit(x_scaled_train, y_train)
```

```python
y_pred = regression.predict(x_scaled_test)
y_pred
```

```python
from sklearn.metrics import r2_score

scrore = r2_score(y_pred,y_test)
```

```python
scrore
```

```python
pickle.dump(scaler, open('scaler.pkl','wb'))
pickle.dump(regression, open('regression.pkl','wb'))
```
